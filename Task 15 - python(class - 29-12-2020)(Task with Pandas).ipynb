```python
"""Task 15 : python(class - 29/12/2020)(Task with Pandas)"""
#Create a DataFrame, save it as a Csv File, set index as 'Random'

import numpy as np
import pandas as pd
```

```python
#np arrays
random = np.random.randint(low = 0, high =100, size = 20)
name = np.random.choice(['A', 'B', 'C', 'D', 'E','F','G','H'],size =  20)
version = np.random.choice([10,11,13,12,14,15,16],size = 20)
gen = np.random.choice([2,3,5,7,9,10],size = 20)
col = np.random.choice(['Black','Orange','Sky','Silver'],size = 20)
bit = np.random.choice([2,4,8,16,32,64,128],size = 20)
#creating dictionary
dict1 = {
    'Random':random, 
    'Name':name,
    'Version':version,
    'Generation':gen,
    'Colour':col,
    'Bit':bit
}

#creating dataframe
df = pd.DataFrame(dict1)
df
```

```python
df.to_csv('file.csv')              # Data Frame converted to Csv file
df.set_index('Random',inplace = True )     #Random Column is set as index
df
```
