```python
"""Task 16 : python(class - 29/12/2020)(Task with Pandas)"""
#Create a Dataframe, and merge it with a new column.

import numpy as np
import pandas as pd
```

```python
store1 = {
    "Pen":['Red','Blue','Black','Orange','Green'],
    "Quantity":[10,20,15,22,38]
}
x1 = pd.DataFrame(store1)
x1
```

```python
store2 = {
    "Pen":['Red','Black','Green','Purple','silver'],
    "Refill Length":[15,12,14,13,11]
}
x2 = pd.DataFrame(store2)
x2
```

```python
xd = pd.merge(x1,x2, on='Pen', how = 'outer')
xd
```
